SALES PREDICTION USING PYTHON

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a synthetic dataset for demonstration
np.random.seed(42) # for reproducibility
data = {
    'TV': np.random.normal(150, 70, 200),
    'Radio': np.random.normal(25, 15, 200),
    'Newspaper': np.random.normal(30, 20, 200),
    'Sales': np.random.normal(15, 5, 200)
}
df = pd.DataFrame(data)

# Ensure values are non-negative and add some correlation for sales
df['TV'] = df['TV'].apply(lambda x: max(0, x))
df['Radio'] = df['Radio'].apply(lambda x: max(0, x))
df['Newspaper'] = df['Newspaper'].apply(lambda x: max(0, x))

# Create a more realistic sales column based on advertising
# Sales = 0.05*TV + 0.2*Radio + 0.01*Newspaper + some_noise + base_sales
df['Sales'] = (0.05 * df['TV'] + 0.2 * df['Radio'] + 0.01 * df['Newspaper'] +
               np.random.normal(0, 2, 200) + 5)
df['Sales'] = df['Sales'].apply(lambda x: max(0, x)) # Ensure non-negative sales

print("First 5 rows of the dataset:")
print(df.head())
print("\nDataset Info:")
df.info()
print("\nDescriptive Statistics:")
print(df.describe())
```

```python

# Pairplot to visualize relationships
sns.pairplot(df, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height=4, aspect=1, kind='reg')
plt.suptitle('Relationship between Advertising Channels and Sales', y=1.02)
plt.show()

# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Advertising Channels and Sales')
plt.show()
```
